```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
from matplotlib import style
style.use('fivethirtyeight') # fivethirtyeight

import plotly.io as pio
pio.renderers
pio.renderers.default = "notebook_connected"
```

```python
dfname = df.groupby(['Name'])
dfname1 = dfname['Name'].count().reset_index(name='Count').sort_values('Count', ascending=False).head(10)
dfname1

fig = px.bar(dfname1, x = 'Count', y = 'Name', title= 'Atheltes who have appeared the most on Forbes highest paid atheletes list', labels={"Name": "Atheltes", "Count": "Number of appearances"})
fig.update_layout(yaxis_categoryorder = 'total ascending')
fig.show()
```

Both Tiger Woods and Michael Jordan have appeared on the list 19 times, which is the highest number of times an athelete has appeared on the list.

```python
money = df.groupby('Name')[['earnings ($ million)']].sum().reset_index().sort_values('earnings ($ million)', ascending=False).head(20)

fig = px.bar(money, x = 'earnings ($ million)', y = 'Name', title= 'Total earnings for Atheletes on the Forbes list between 1990-2020', labels={"Name": "Athletes", "earnings ($ million)": "Total earnings (in US$ million)"})
fig.update_layout(yaxis_categoryorder = 'total ascending')
fig.show()
```

tiger Woods is on top again, he has earned the most and also appeared the most in forbes list. While Michael Jordan has appeared as many times as Tiger Woods, he has earned about $500 Million less than Tiger Woods in his career in forbes list.

# Which countries have the high earning athletes and how much have they earned?

```python
countries = dfname = df.groupby(['Nationality'])
nationality = dfname['Nationality'].count().reset_index(name='Count').sort_values('Count', ascending=False).head(10)

fig = px.bar(nationality, x = 'Count', y = 'Nationality', title= 'Amount of times atheltes from these countries have appeared on the list')
fig.update_layout(yaxis_categoryorder = 'total ascending')
fig.show()
```

```python
#Fixing the names of the sport to group them together
df['Sport'] = df['Sport'].str.capitalize()


sport = df.groupby('Sport')[['earnings ($ million)']].sum().reset_index().sort_values('earnings ($ million)', ascending=False).head(15)

fig = px.bar(sport, x = 'earnings ($ million)', y = 'Sport', title= 'Total earnings for Atheletes on the Forbes list between 1990-2020', labels={"Name": "Athletes", "earnings ($ million)": "Total earnings (in US$ million)"})
fig.update_layout(yaxis_categoryorder = 'total ascending')
fig.show()
```

the highest earning game was basketball.
